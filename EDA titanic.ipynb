```python
# select and import dataset

import pandas as pd
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
# data view

print(df.head())
print(df.info())
print(df.describe())
```

```python
# data cleansing
print(df.isnull().sum())
df= df.dropna()
```

```python
# distribution of numerical variables

import matplotlib.pyplot as plt
plt.figure (figsize = (8,4))
sns.histplot (df['age'], bins = 30)
plt.title ('Distribution age')
plt.show()

# frequency of categorical variables

sns.countplot (x= 'class', data = df)
plt.title ('Distribution classes sociales')
plt.show()
```

```python
# correlation between numeric values

sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm")
plt.title ('Correlation matrix')
plt.show()
```

```python
# relation between two numeric values

sns.scatterplot(x='age', y='fare', data=df)
plt.title('Relation btw age and fare')
plt.show()
```

```python
# comparing a numeric variable btw many categories

sns.boxplot(x='class', y='age', data=df)
plt.title('Class partition according to age')
plt.show()
```

```python
# survival rate and gender according to gender

survival_rate_by_gender = df.groupby('sex')['survived'].mean()
print(survival_rate_by_gender)

survival_num_by_gender = df.groupby('sex')['survived'].count()
print(survival_num_by_gender)
```
